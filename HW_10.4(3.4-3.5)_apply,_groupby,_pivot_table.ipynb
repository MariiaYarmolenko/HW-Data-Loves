1. –î–æ—Å–ª—ñ–¥–∏–º–æ, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª —Å—É–ø–µ—Ä–º–∞—Ä–∫–µ—Ç—É ('Branch') —î –Ω–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à–∏–º. –î–ª—è —Ü—å–æ–≥–æ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑–∞ –∫–æ–∂–Ω–∏–º —Ñ—ñ–ª—ñ–∞–ª–æ–º —ñ –≤–∏—è–≤—ñ—Ç—å, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª –º–∞—î –Ω–∞–π–≤–∏—â–∏–π.

```python
branch_profit = df.groupby('Branch')['gross income'].sum()
print(branch_profit)

top_branch = branch_profit.idxmax()
max_profit = branch_profit.max()
print(top_branch)
print(max_profit)
```

2. –í —è–∫–æ–º—É –º—ñ—Å—Ç—ñ –∑–Ω–∞–π—Ö–æ–¥–∏—Ç—å—Å—è —Ñ—ñ–ª—ñ–∞–ª –∑ –Ω–∞–π–≤–∏—â–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º? –ú–æ–∂–µ –≤ —Ç–æ–º—É –º—ñ—Å—Ü—ñ –Ω–∞–º —Ä–æ–∑–º—ñ—Ç–∏—Å—Ç–∏ —â–µ –æ–¥–∏–Ω –º–∞–≥–∞–∑–∏–Ω.  
–ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—å –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É Pandas.

```python
city_profit = df.groupby(['City', 'Branch'])['gross income'].sum()
top_city = city_profit.idxmax()
max_profit_city = city_profit.max()
print(top_city)
print(max_profit_city)
```

3.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –∑–≤–æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—é, —è–∫–∞ –ø–æ–∫–∞–∂–µ, —Å–∫—ñ–ª—å–∫–∏ –ø–æ–∫—É–ø–æ–∫ (—ñ–Ω–≤–æ–π—Å—ñ–≤) –±—É–ª–æ –∑—Ä–æ–±–ª–µ–Ω–æ –≤ –∫–æ–∂–Ω—ñ–π –∑ —Ñ—ñ–ª—ñ–π (`Branch`) –∑–∞ —Ä—ñ–∑–Ω–∏–º–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è–º–∏ —Ç–æ–≤–∞—Ä—ñ–≤. –ó–∞–ø–∏—à—ñ—Ç—å —Ç–∞–±–ª–∏—Ü—é –≤ –∑–º—ñ–Ω–Ω—É `invoices_by_category` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –∑–º—ñ–Ω–Ω—É –Ω–∞ –µ–∫—Ä–∞–Ω.
–¶—è —Ç–∞–±–ª–∏—Ü—è –¥–æ–ø–æ–º–æ–∂–µ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –≤ —è–∫—ñ–π —Ñ—ñ–ª—ñ—ó –∫—É–ø—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∫–æ–∂–Ω–æ—ó –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.

```python
invoices_by_category = pd.pivot_table(df, index='Branch', columns='Product line', values='Invoice ID', aggfunc='size') #–ê–±–æ count
display(invoices_by_category)
```

```python
max_invoices_for_electronics = invoices_by_category['Electronic accessories'].idxmax()
max_value_for_electronics = invoices_by_category['Electronic accessories'].max()
print(max_invoices_for_electronics)
print(max_value_for_electronics)
```

```python
loyalty_analysis = df.groupby('Customer type').agg({
    'Rating':'mean',
    'Total':'mean',
    'Quantity':'sum'
})
display(loyalty_analysis)
```

2. –ì—ñ–ø–æ—Ç–µ–∑–∞ –ø–ø—Ä–æ –∑–≤'—è–∑–æ–∫ —Ä–µ–π—Ç–∏–Ω–≥—É —Ç–∞ —Å—É–º–∏ —á–µ–∫—É.

–ü–∏—Ç–∞–Ω–Ω—è: —á–∏ —ñ—Å–Ω—É—î –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å –º—ñ–∂ —Å—É–º–æ—é —á–µ–∫–∞ (Total) —Ç–∞ —Ä–µ–π—Ç–∏–Ω–≥–æ–º (Rating)?

–ß–æ–º—É —Ü–µ –≤–∞–∂–ª–∏–≤–æ: –ø–æ–∫—É–ø—Ü—ñ, —è–∫—ñ –≤–∏—Ç—Ä–∞—á–∞—é—Ç—å –±–∞–≥–∞—Ç–æ –≥—Ä–æ—à–µ–π, –º–∞—é—Ç—å –≤–∏—â—ñ –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è –≤—ñ–¥ —Å–µ—Ä–≤—ñ—Å—É. –Ø–∫—â–æ –≤–µ–ª–∏–∫—ñ —á–µ–∫–∏ —Å—É–ø—Ä–æ–≤–æ–¥–∂—É—é—Ç—å—Å—è –Ω–∏–∑—å–∫–∏–º–∏ –æ—Ü—ñ–Ω–∫–∞–º–∏, —Ü–µ —Å–∏–≥–Ω–∞–ª, —â–æ –ø—Ä–µ–º—ñ—É–º-—Å–µ–≥–º–µ–Ω—Ç –∞–±–æ –ø–æ—Å—Ç—ñ–π–Ω—ñ –ø–æ–∫—É–ø—Ü—ñ –∑ —à–∏—Ä–æ–∫–∏–º–∏ –∫–æ—à–∏–∫–∞–º–∏ –Ω–µ–∑–∞–¥–æ–≤–æ–ª–µ–Ω–∏–π —Ä—ñ–≤–Ω–µ–º –æ–±—Å–ª—É–≥–æ–≤—É–≤–∞–Ω–Ω—è.

–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—è: –≤ –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –≤—ñ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤, —è–∫—â–æ –ø—ñ–¥—Ç–≤–µ—Ä–¥–∏—Ç—å—Å—è –∑–≤–æ—Ä–æ—Ç–Ω—ñ–π –∑–≤'—è–∑–æ–∫, –∑–∞–ø—Ä–æ–≤–∞–¥–∏—Ç–∏ –¥–æ–¥–∞—Ç–∫–æ–≤—ñ –ø–æ—Å–ª—É–≥–∏ –¥–ª—è "–≤–µ–ª–∏–∫–∏—Ö —á–µ–∫—ñ–≤", —è–∫—ñ –ø–æ–ª—ñ–ø—à–∞—Ç—å –∑–∞–¥–æ–≤–æ–ª–µ–Ω—ñ—Å—Ç—å –∫–ª—ñ—î–Ω—Ç—ñ–≤ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –¥–æ–ø–æ–º–æ–≥–∞ –∑ –¥–æ—Å—Ç–∞–≤–∫–æ—é –¥–æ –º–∞—à–∏–Ω–∏/–¥–æ –¥–æ–º—É –¥–ª—è –≥–∞–±–∞—Ä–∏—Ç–Ω–∏—Ö –ø–æ–∫—É–ø–æ–∫; –æ–∫—Ä–µ–º—ñ –∫–∞—Å–∏ –∑ –ø–∞–∫—É–≤–∞–Ω–Ω—è–º —Ç–æ—â–æ)

```python
def segment_total(total):
  if total < 200:
    return 'Small'
  elif total < 600:
    return 'Medium'
  else:
    return 'Large'

df['Total Size'] = df['Total'].apply(segment_total)
satisfaction_by_check = df.groupby('Total Size')['Rating'].mean()
print(satisfaction_by_check)
```
